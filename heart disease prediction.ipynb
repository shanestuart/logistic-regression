```python
cp=int(input("Enter the chest pain type:"))
thal=int(input("Enter the taltha:"))
ca=int(input("Enter the number of major vessels:"))
thalach=int(input("Enter the thalach: "))
while True:
    try:
        oldpeak_input = input("Enter the oldpeak:")
        oldpeak = float(oldpeak_input)
        break
    except ValueError:
        print("Invalid input. P1lease enter a numerical value for oldpeak.")
pred=Logr.predict([[cp,thal,ca,thalach,oldpeak]])
print(pred)
```
